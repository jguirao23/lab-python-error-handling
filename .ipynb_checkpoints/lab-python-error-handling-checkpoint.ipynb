def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"] 
inventory = {}

def initialize_inventory(products): 
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory

def get_customer_orders():
    print("\n")
    customer_orders = set()
    order_by_customer = 0
    valid_number_customer_orders = False
    while not valid_number_customer_orders:
        try:
            number_customer_orders = int(input("Enter the number of customer orders: "))
            if number_customer_orders > 0:
                valid_number_customer_orders = True
            else:
                print("The number of orders cannot be negative. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a valid number of orders")    
    for orders in range(number_customer_orders):
        valid_customer_order = False
        while not valid_customer_order:
            try:
                order_by_customer = input("Enter the name of a product that you want to order: ")
                if order_by_customer in inventory.keys() and inventory[order_by_customer] > 0 :
                    customer_orders.add(order_by_customer)
                    valid_customer_order = True
                else: 
                    print("The product is not available.")
            except ValueError:
                print("Invalid order. Please enter a valid product.")
    return customer_orders

def total_price_customer_order(customer_orders):
    print("\n")
    prices = {}
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                product_price = int(input("Enter the price of " + product + ": "))
                if product_price >= 0:
                    prices[product] = product_price
                    valid_price = True
                else:
                    print(f"Invalid input. Please enter a valid price for {product}.")
            except ValueError:
                print(f"Invalid input. Please enter a valid price for {product}.")
    total_price = sum(prices.values())
    return total_price

def calculate_order_statistics(customer_orders, products):
    print("\n")
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered) / len(inventory)
    percentage_ordered = "{:.2%}".format(percentage_ordered)
    order_statistics = (total_products_ordered, percentage_ordered)
    print("Order statistics:")
    print(f"Total products ordered: {total_products_ordered}") 
    print(f"Percentage of unique products ordered: {percentage_ordered}") 
    
def update_inventory(customer_orders, inventory):
    inventory = {key:(value-1 if key in list(customer_orders) else value) for key,value in inventory.items()}
    updated_inventory = {key:value for key,value in inventory.items() if value > 0}
    
    return updated_inventory
````

````python
inventory = initialize_inventory(products) 

customer_orders = get_customer_orders()

calculate_order_statistics(customer_orders, products)

inventory = update_inventory(customer_orders, inventory)

print("Updated inventory:")
for key,value in inventory.items():
    print(f"{key}: {value}")

print(f"Total price is {total_price_customer_order(customer_orders)}")
````
